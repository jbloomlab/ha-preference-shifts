```python
from Bio.PDB import PDBParser, PDBIO, Select

def extract_chain(input_pdb, output_pdb, chain_ids, model_id=1):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure("structure", input_pdb)
    
    class ChainSelect(Select):
        def accept_model(self, model):
            return model.get_id() == model_id
        
        def accept_chain(self, chain):
            return chain.get_id() in chain_ids
    
    io = PDBIO()
    io.set_structure(structure)
    io.save(output_pdb, ChainSelect())

extract_chain("../structures/pdbs/4o5n.pdb", "../structures/rewritten_pdbs/4o5n_foldmason_A.pdb", ["A"])
extract_chain("../structures/pdbs/4o5n.pdb", "../structures/rewritten_pdbs/4o5n_foldmason_B.pdb", ["B"])

extract_chain("../structures/pdbs/4kwm.pdb", "../structures/rewritten_pdbs/4kwm_foldmason_A.pdb", ["A"])
extract_chain("../structures/pdbs/4kwm.pdb", "../structures/rewritten_pdbs/4kwm_foldmason_B.pdb", ["B"])

extract_chain("../structures/pdbs/4r8w.pdb", "../structures/rewritten_pdbs/4r8w_foldmason_A.pdb", ["A"])
extract_chain("../structures/pdbs/4r8w.pdb", "../structures/rewritten_pdbs/4r8w_foldmason_B.pdb", ["B"])
```
