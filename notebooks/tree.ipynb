```python
import pandas as pd
import numpy as np
import altair as alt

from Bio import AlignIO, Phylo
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor

from itertools import combinations
from natsort import natsort_keygen
import theme
```

```python
aln = AlignIO.read("../data/ha18_alignment_revised.fasta", "fasta")

# note this identity distance includes gaps, more accurate pairwise identity
# excludes gaps and is calculated in sequence_identity.ipynb
calc = DistanceCalculator("identity")
dm = calc.get_distance(aln)
tree = DistanceTreeConstructor().nj(dm)

# Remove names from internal nodes
for clade in tree.get_nonterminals():
    clade.name = None

Phylo.write(tree, "../results/ha18_tree/ha18.newick", "newick")
Phylo.draw_ascii(tree)
```
